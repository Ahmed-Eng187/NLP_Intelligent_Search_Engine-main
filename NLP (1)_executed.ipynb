### 10. Text Preprocessing

**What this step does:**
1.  Downloads additional NLTK resources (`punkt`, `punkt_tab`).
2.  Initializes a set of English stopwords.
3.  Defines a `preprocess` function that performs the following steps on input text:
    *   Converts the text to lowercase.
    *   Removes all non-alphabetic characters using regular expressions.
    *   Tokenizes the text into individual words using `word_tokenize`.
    *   Removes common English stop words (e.g., 'the', 'is', 'a').
    *   Joins the cleaned tokens back into a single string.
4.  Applies this `preprocess` function to the `text` column of the DataFrame to create a new `clean_text` column.

**Why this step is important:** Text preprocessing is a fundamental step in almost all NLP tasks. It cleans and normalizes the text data, reducing noise and ensuring consistency. Lowercasing prevents treating 'Apple' and 'apple' as different words, removing special characters eliminates irrelevant symbols, tokenization breaks text into analyzable units, and stop word removal focuses on more meaningful terms, all of which significantly improve the quality of text representations for search and similarity calculations.

**Technique/Method:** Text preprocessing (lower-casing, regular expressions, NLTK tokenization, stop word removal).

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

# تجهيز stopwords
stop_words = set(stopwords.words('english'))

# دالة التنضيف
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)

# تطبيق التنضيف
df['clean_text'] = df['text'].apply(preprocess)

# عرض
df[['text', 'clean_text']].head()
```

### 14. TF-IDF Feature Extraction

**What this step does:** Initializes `TfidfVectorizer` (with n-grams and `min_df`) and transforms `clean_text` into a TF-IDF matrix.

**Why this step is important:** Converts text into numerical vectors, capturing word importance and context for machine learning.

**Technique/Method:** Feature extraction using TF-IDF (Term Frequency-Inverse Document Frequency).

### 14. TF-IDF Feature Extraction

**What this step does:**
1.  Initializes a `TfidfVectorizer` from `sklearn.feature_extraction.text`.
    *   `ngram_range=(1,2)`: Configures the vectorizer to consider both single words (unigrams) and two-word phrases (bigrams).
    *   `min_df=2`: Excludes terms that appear in fewer than 2 documents, filtering out very rare words.
2.  Transforms the `clean_text` column of the DataFrame into a TF-IDF (Term Frequency-Inverse Document Frequency) matrix.
3.  Prints the shape of the resulting `tfidf_matrix`.

**Why this step is important:** TF-IDF is a powerful statistical measure used to evaluate how important a word is to a document in a collection. It converts raw text into a numerical matrix, which is a format required by machine learning algorithms. By considering n-grams, it captures more contextual information, and `min_df` helps to reduce noise from infrequent terms, leading to more robust text representations.

**Technique/Method:** Feature extraction using TF-IDF (Term Frequency-Inverse Document Frequency).

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1,2), min_df=2)
tfidf_matrix = vectorizer.fit_transform(df['clean_text'])

print(tfidf_matrix.shape)
```

### 15. TF-IDF Search Function

**What this step does:** Defines a function `search` that preprocesses a query, converts it to a TF-IDF vector, calculates cosine similarity with document vectors, and returns top `k` similar products.

**Why this step is important:** Implements the core TF-IDF recommendation logic, finding relevant products based on textual similarity.

**Technique/Method:** Information retrieval, TF-IDF, Cosine Similarity, Top-K retrieval.

### 15. TF-IDF Search Function

**What this step does:** Defines a function `search` that performs a product recommendation using TF-IDF and Cosine Similarity.
1.  The input `query` is preprocessed using the earlier defined `preprocess` function.
2.  The preprocessed `query` is transformed into a TF-IDF vector using the trained `vectorizer`.
3.  Cosine similarity is calculated between the `query_vec` and all document vectors in `tfidf_matrix`.
4.  The indices of the top `k` most similar products are identified.
5.  The function returns a DataFrame containing these top `k` products along with their similarity scores.

**Why this step is important:** This function implements the core logic for the TF-IDF-based recommendation system. Cosine similarity is a common metric for determining the similarity between two non-zero vectors in an inner product space, effectively measuring the 'angle' between text representations. This allows the system to retrieve products most relevant to a user's textual query.

**Technique/Method:** Information retrieval, text representation using TF-IDF, similarity calculation using Cosine Similarity, Top-K retrieval.

```python
from sklearn.metrics.pairwise import cosine_similarity

def search(query, top_k=5):
    query = preprocess(query)

    query_vec = vectorizer.transform([query])

    similarity = cosine_similarity(query_vec, tfidf_matrix)[0]

    top_indices = similarity.argsort()[::-1][:top_k]

    results = df.iloc[top_indices].copy()
    results['score'] = similarity[top_indices]

    return results
```

### 19. TF-IDF Evaluation Function (Precision@K)

**What this step does:** Defines `evaluate_tfidf` to search, count matches with an `expected_category`, and calculate Precision@K.

**Why this step is important:** Quantitatively assesses the TF-IDF model's effectiveness using a standard metric.

**Technique/Method:** Evaluation metric: Precision@K.

### 19. TF-IDF Evaluation Function (Precision@K)

**What this step does:** Defines an evaluation function `evaluate_tfidf` that:
1.  Executes the TF-IDF `search` function with a given `query` and `top_k` value.
2.  Iterates through the `Bb Category` of the retrieved top `k` results.
3.  Counts how many of these categories match the `expected_category` (case-insensitively).
4.  Calculates Precision@K (the fraction of relevant items among the top `k` retrieved items) and prints the result.
5.  Returns the `results` DataFrame.

**Why this step is important:** This function provides a quantitative metric (`Precision@K`) to objectively assess the effectiveness of the TF-IDF search model. By automating the evaluation process, it allows for consistent and reproducible measurement of how accurately the model retrieves products belonging to a desired category for a given query.

**Technique/Method:** Evaluation metric: Precision@K.

```python
def evaluate_tfidf(query, expected_category, top_k=5):
    results = search(query, top_k)

    count = 0
    for cat in results['Bb Category']:
        if expected_category.lower() in cat.lower():
            count += 1

    precision = count / top_k
    print("TF-IDF Precision@", top_k, "=", precision)

    return results
```

### 22. BERT Search Function

**What this step does:** Defines `search_bert` to preprocess a query, encode it into a BERT embedding, calculate cosine similarity with document embeddings, and return top `k` similar products.

### 22. BERT Search Function

**What this step does:** Defines a function `search_bert` that performs product recommendations using BERT embeddings and Cosine Similarity.
1.  The input `query` is first preprocessed using the `preprocess` function.
2.  The preprocessed `query` is then transformed into a BERT embedding using the `SentenceTransformer` model.
3.  Cosine similarity is calculated between this `query_embedding` and all pre-computed `doc_embeddings`.
4.  The indices of the top `k` most similar products are retrieved.
5.  The function returns a DataFrame containing these top `k` products along with their similarity scores.

**Why this step is important:** This function implements the core logic for the BERT-based recommendation system. By leveraging the semantic understanding of BERT to generate embeddings, it can identify more contextually relevant products, even if the query terms don't explicitly appear in the product descriptions. Cosine similarity is used to find the most semantically close items.

**Technique/Method:** Information retrieval, text representation using BERT embeddings, similarity calculation using Cosine Similarity, Top-K retrieval.

```python
from sklearn.metrics.pairwise import cosine_similarity

def search_bert(query, top_k=5):
    query = preprocess(query)

    query_embedding = model.encode([query])

    similarity = cosine_similarity(query_embedding, doc_embeddings)[0]

    top_indices = similarity.argsort()[::-1][:top_k]

    results = df.iloc[top_indices].copy()
    results['score'] = similarity[top_indices]

    return results
```

### 26. BERT Evaluation Function (Precision@K)

**What this step does:** Defines an evaluation function `evaluate_bert` that:
1.  Executes the BERT-based `search_bert` function with a given `query` and `top_k` value.
2.  Iterates through the `Bb Category` of the retrieved top `k` results.
3.  Counts how many of these categories match the `expected_category` (case-insensitively).
4.  Calculates Precision@K (the fraction of relevant items among the top `k` retrieved items) and prints the result.
5.  Returns the `results` DataFrame.

**Why this step is important:** Similar to the TF-IDF evaluation function, this provides a quantitative metric (`Precision@K`) to objectively assess the effectiveness of the BERT-based semantic search model. It's crucial for comparing BERT's performance against TF-IDF using a standardized evaluation measure.

**Technique/Method:** Evaluation metric: Precision@K.

```python
def evaluate_bert(query, expected_category, top_k=5):
    results = search_bert(query, top_k)

    count = 0
    for cat in results['Bb Category']:
        if expected_category.lower() in cat.lower():
            count += 1

    precision = count / top_k
    print("BERT Precision@", top_k, "=", precision)

    return results
```
